```python
im0 = cv2.imread("data/sequence/frames0_0.png")
size = (im0.shape[1], im0.shape[0])
stereo = cv2.stereoRectify(c['K0'], c['d0'], c['K1'],
c['d1'], size, c['R'], c['t'], flags=0)
R0, R1, P0, P1 = stereo[:4]
maps0 = cv2.initUndistortRectifyMap(c['K0'], c['d0'], R0, P0, size, cv2.CV_32FC2)
maps1 = cv2.initUndistortRectifyMap(c['K1'], c['d1'], R1, P1, size, cv2.CV_32FC2)
```

You can now rectify images by doing cv2.remap(im, *maps, cv2.INTER_LINEAR), where maps
is either maps0 or maps1 depending on which camera the image is coming from. This also handles
the undistortion of the images. P0 and P1 are the projection matrices for the rectified images,
which will come in handy when we need to triangulate the points in 3D.

```python
# Rectify images
ims0 = []
ims1 = []

for i in range(26):
    # image0 = cv2.imread(f"data/sequence/frames0_{i}.png", cv2.IMREAD_GRAYSCALE) / 255
    image0 = cv2.imread(f"data/sequence/frames0_{i}.png") / 255
    image0 = np.mean(image0, axis=2) 
    # image1 = cv2.imread(f"data/sequence/frames1_{i}.png", cv2.IMREAD_GRAYSCALE) / 255
    image1 = cv2.imread(f"data/sequence/frames1_{i}.png") / 255
    image1 = np.mean(image1, axis=2) 
    ims0.append(cv2.remap(image0, *maps0, cv2.INTER_LINEAR))
    ims1.append(cv2.remap(image1, *maps1, cv2.INTER_LINEAR))

ims0 = np.array(ims0)
ims1 = np.array(ims1)

ims0.shape, ims1.shape
```

# 13.3

```python
def unwrap(ims, n1:int=40):
    
    n2 = n1+1
   
    # Get the images with pattern 1 (40 periods). The First to images are the basis images. 
    primary_images = ims[2:18]
    # Get the images with pattern 2 (41 periods)
    secondary_images = ims[18:]
    
    #  Apply fast Fourier transofrom to obtain the phase (frequency of the pattern)
    fft_primary = np.fft.rfft(primary_images, axis=0)
    # Apply FFT to secondary 
    fft_secondary = np.fft.rfft(secondary_images, axis=0)
       
    # Obtain the angles of the phase, theta. Which is the second element of the FFT. That is a comples number 
    theta_primary = np.angle(fft_primary[1])
    theta_secondary = np.angle(fft_secondary[1])
    
    # We have decided that n2 = n1 + 1.
    # We use the Heterodyne principle to calculate the phase cue 
    theta_c = np.mod(theta_secondary-theta_primary, 2*np.pi)
       
    # We calculate how many times theta_primary has wraped around, the rounding makes it robust to noise 
    o_primary = np.round(((n1 * theta_c) - theta_primary) / (2*np.pi), 100)
        
    # Use the primary phase to obtain the unwrapped phase (theta).
    theta_est = (((2 * np.pi * o_primary) + theta_primary) / n1 )
    
    return theta_est    
```

```python
theta_0 = unwrap(ims0)
theta_1 = unwrap(ims1)
```

```python
fig, ax = plt.subplots(1,2, figsize=(10,10))
ax[0].imshow(theta_0, cmap="inferno")
ax[1].imshow(theta_1, cmap="inferno")
```

# 13.4

When inspecting the phase images theta0 and theta1 it is clear that not all pixels contain a
valid measurement of the phase. This is because some pixels do not reflect enough light from the
projector to give a meaningful measurement.

```python
# Set a trheshold 
tr = 15 / 255

# For each camera we have to create a mask to obtain the pixels that are bright enough. 
diff0 = ims0[0] - ims0[1] 
diff1 = ims1[0] - ims1[1]

mask0 = diff0 > tr 
mask1 = diff1 > tr 

fig, ax = plt.subplots(1,2, figsize=(10,10))

ax[0].imshow(mask0, cmap='gray')
ax[1].imshow(mask1, cmap='gray')
```

```python
q0s = []
q1s = []

rows0, cols0 = mask0.shape
rows1, cols1 = mask1.shape
# Iterate through the rows, rectified image, epipolar lines lie in the same row
for r0 in range(rows0):
    # Iterate through columns
    for c0 in range(cols0):
        # check if the pisxels is valid (if it has value on the mask) then map the phase 
        if mask0[r0, c0]: 
            # set the value of theta 1
            theta0_val = theta_0[r0, c0]
            # set initial values for theta 1
            theta1_val = np.inf
            theta1_val_index = 0
            for c1 in range(cols1):
                # check that the value has a value in the mask and has the closest phase match 
                # phase match: min -> |theta_0-theta_1| if it's equal means that the phase are the same 
                if mask1[r0, c1] and (np.abs(theta_1[r0, c1] - theta0_val)) < theta1_val:
                    # Update the value of phase match
                    theta1_val = np.abs(theta_1[r0, c1] - theta0_val)
                    # store the index of the column 
                    theta1_val_index = c1
                    # Iterate through all the columns, store the min value 

            # Store the pixel coordinate 
            q0s.append([c0, r0])
            # Store the phase match and the row 
            q1s.append([theta1_val_index, r0])
                    
q0s = np.array(q0s)
q1s = np.array(q1s)

q0s.shape, q1s.shape
```

```python
# Initialize the disparity with all zeros 
disparity = np.zeros(mask0.shape)
# calculate disparity for each pixel 
for (j0, i0), (j1, i1) in zip(q0s, q1s):
    disparity[i0, j0] = j0 - j1
```

```python
disparity = cv2.medianBlur(disparity.astype(np.float32), 5)
fig, ax = plt.subplots(figsize=(10,10))
ax.imshow(disparity, cmap="inferno")
```

```python
def inhom(Q):
    """
    Returns the Point/s Q in inhomogenous form. 
    Given Q divides by the scale:
        - 2D points:
            [sx, sy, s]/s = [x, y, 1] and we return [x, y]
        - 3D points:
            As the scale is 1 we divide by one and return everything but the last column.
    
    ARGUMENTS
    ---------
    Q -> Point or set of points in the form of column vector. (3, n) or in case of 3d points (4,n)
    
    RETURNS
    -------
    Q_tmp -> Points in inhomogenous form. 2d: (2,n), 3d: ()
    """
    # Get the scale 
    s = Q[-1,:]
    
    # Divide by the scale at select everything but the last row
    Q_tmp = (Q/s)[:-1,:]
    
    return Q_tmp
```

```python
points0 = np.array(q0s).astype(np.float32).T
points1 = np.array(q1s).astype(np.float32).T
points0.shape, points1.shape
```

```python
points3d = cv2.triangulatePoints(P0, P1, points0, points1)
points3d.shape
```

```python
points3d_inh = inhom(points3d)
points3d_inh.shape
```

```python
color = []

for (j0, i0), (j1, i1) in zip(q0s, q1s):
    color.append(j0 - j1)

color = np.array(color)
color = cv2.medianBlur(color.astype(np.float32), 5)

cmap = matplotlib.cm.get_cmap('inferno')
norm = matplotlib.colors.Normalize(vmin=color.min(), vmax=color.max())
rgb_colors = []

for c in color:
    rgb_colors.append(cmap(norm(c))[0, :3])

rgb_colors = np.array(rgb_colors)
```

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(projection='3d')

ax.scatter(points3d_inh[1], points3d_inh[0], points3d_inh[2],
           c=rgb_colors, s=1)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_xlim([-300, 300])
ax.set_ylim([-100, 500])
ax.set_zlim([400, 1000])
ax.view_init(75, -5)
# ax.azim = 10
# ax.dist = 10
# ax.elev = 10

plt.show()
```
